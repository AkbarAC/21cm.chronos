```python
z = np.arange(6, 16, 2)
print(z)
c = ['b', 'g', 'y', 'orange', 'r']

zs = np.arange(5, 15, 0.5)
```

```python
#Brightness temperature function
def brightness_temperature(box, redshift, eff=None, vir=None,xray=None,path=None):
    initial_conditions = p21c.initial_conditions(
        user_params = {"HII_DIM": box, "BOX_LEN": box},
        cosmo_params = p21c.CosmoParams(SIGMA_8=0.8),
        random_seed=54321
    )

    perturbed_field = p21c.perturb_field(
        redshift = redshift,
        init_boxes = initial_conditions
    )

    ionized_field = p21c.ionize_box(
        perturbed_field = perturbed_field,
        astro_params = p21c.AstroParams(HII_EFF_FACTOR=eff, ION_Tvir_MIN=vir),
    )

    brightness_temp = p21c.brightness_temperature(ionized_box=ionized_field, perturbed_field=perturbed_field)

    return brightness_temp
```

```python
def ps1d(cube, kbins=15):
    H = 67.66  
    h = H/100.
    dim = 100.

    box_dims = [dim]*len(cube.shape)

    ft = fftpack.fftshift(fftpack.fftn(cube.astype('float64')))
    
    power_spectrum = np.abs(ft)**2
    
    boxvol = np.product(box_dims)
    pixelsize = boxvol/(np.product(cube.shape))
    
    power_spectrum *= pixelsize**2/boxvol

    ps, ks, n_modes = radial_average(power_spectrum, kbins=kbins, box_dims=box_dims, binning='log', breakpoint=0.1)
    return ks, ps*ks**3/2/pi**2
```

```python
Yf = np.array([brightness_temperature(box=50, redshift=i).brightness_temp for i in z])
Ye = np.array([brightness_temperature(box=50, redshift=i, eff=35.0).brightness_temp for i in z])
Yt = np.array([brightness_temperature(box=50, redshift=i, vir=5.10).brightness_temp for i in z])
Ya = np.array([brightness_temperature(box=50, redshift=i,eff=35.0,vir=5.10).brightness_temp for i in z])
```

```python
scale_Yf, power_Yf = [], []

for i,ii in enumerate(z):
    ks, ps = ps1d(Yf[i,...])
    scale_Yf.append(ks); power_Yf.append(ps)
    
scale_Ye, power_Ye = [], []

for i,ii in enumerate(z):
    ks, ps = ps1d(Ye[i,...])
    scale_Ye.append(ks); power_Ye.append(ps)
    
scale_Yt, power_Yt = [], []

for i,ii in enumerate(z):
    ks, ps = ps1d(Yt[i,...])
    scale_Yt.append(ks); power_Yt.append(ps)

    
scale_Ya, power_Ya = [], []

for i,ii in enumerate(z):
    ks, ps = ps1d(Ya[i,...])
    scale_Ya.append(ks); power_Ya.append(ps)
```

```python
Yfs = np.array([brightness_temperature(box=50, redshift=i).brightness_temp for i in zs])
Yes = np.array([brightness_temperature(box=50, redshift=i, eff=40.0).brightness_temp for i in zs])
Yts = np.array([brightness_temperature(box=50, redshift=i, vir=5.10).brightness_temp for i in zs])
Yas = np.array([brightness_temperature(box=50, redshift=i,eff=40.0,vir=5.10).brightness_temp for i in zs])
```

```python
plt.plot(zs, np.std(Yfs, axis=(1,2,3)), '-', label='$Y_f$')
plt.plot(zs, np.std(Yes, axis=(1,2,3)), 'o--', label='$Y_\epsilon$')
plt.plot(zs, np.std(Yts, axis=(1,2,3)), 'v--', label='$Y_T$')
plt.plot(zs, np.std(Yas, axis=(1,2,3)), '^--', label='$Y_A^*$')


plt.xlabel('Redshift $z$')
plt.ylabel('Brightness temperature of HI [mK]')
plt.legend()
plt.grid()
plt.savefig('SpectrumAstroparam.pdf')
```

```python
fig = plt.figure(figsize=(12,4))
gs = fig.add_gridspec(1, 4, hspace=0, wspace=0)
ax = gs.subplots(sharex=True, sharey=True)

scale, power = [scale_Yf, scale_Ye, scale_Yt, scale_Ya], [power_Yf, power_Ye, power_Yt, power_Ya]
titles = ['$Y_f$', '$Y_\epsilon$', '$Y_T$','$Y_A^*$']

for i, ii in enumerate(z):
    for j,jj in enumerate(ax):
        jj.loglog(scale[j][i], power[j][i], '-', color=c[i], label='z = %i'%z[i])
        jj.set_xlabel('$k$ [Mpc$^{-1}$]')
        jj.grid(True, which='major', color='grey')
        jj.grid(True, which='minor')
        if j==2: jj.legend(ncol=2)
        jj.set_title(titles[j], ha='center');
        
ax[0].set_ylabel('$k^3/(2\pi^2)\ P(k)$ [mK$^2$ Mpc$^{-3}$]')
plt.savefig('PSvskAstroparams.pdf')
```
